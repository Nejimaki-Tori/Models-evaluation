```python
hint = """Given a query response and a sentence from that response, determine
if that sentence causes any confusion. Types of confusion include
the following:
- Causal omission: the reason or motivation for something is
missing or unclear
- Salience: inclusion of trivial details that do not contribute to
the main storyline
- Discontinuity: an interruption in the flow of the narrative,
including but not restricted to: sudden jumps between
perspectives, time periods, or settings; poor transition between
sentences; sentences that seem out of
place; illogical sentence order of query response structure
- Duplication: redundant repetition of similar information
- Inconsistency: two parts of the query response contain contradicting
information
- Language: grammar issues; confusing wording or phrasing; etc.
For something to qualify as a confusion, it must meet these two
conditions:
1. Without resolving the confusion, readers would struggle
substantially to grasp the main narrative, or the query response would
appear incoherent.
2. The confusion can’t be resolved solely using the information
provided in the query response or the response doesn't correspond to the query.
If a given sentence contains serious confusion that corresponds to these two,
ask relevant clarifying questions and provide the confusion types. There can be multiple questions, and keep
in mind that a sentence may involve multiple types of confusion.
If you don’t identify any severe mistakes and serious confusion like that 
please say "no confusion" for both questions and types in your response.
Some examples are provided below, followed by the
query response and sentence that you need to annotate.
Please, annotate only the given sentence at the beggining and not the whole text.

[Example response]
Questions: no confusion
Types: no confusion
"""

task = """
Determine if the given sentence contains serious confusion that cannot be resolved by any 
part of the given query response, or if it does not correspond to the user's request.
If you don’t identify any confusion like that, simply say "no confusion" for both questions and types in your response.
"""

def proc_batch(batch, name, text, max_retries=10):
    global client
    while len(batch) > 0 and max_retries > 0:
        jobs = []
        for idx, context in batch:
            job = None
            job = client.submit(
                task=task,
                context="Sentence: " + context + "\n\n\n\n" + text,
                hint=hint,
                temperature=0.01,
                beam_size=1,
                repetition_penalty=1,
                model_name=name,
                language="russian",
                force_language=False,
                api_name="/http_bot"
            )
            jobs.append((idx, job, context))
            # to prevent race condition
            time.sleep(0.3)
        new_batch = []
        for j, (idx, job, context) in enumerate(jobs):
            try:
                yield (idx, job.result(timeout=100), context)
            except Exception as err:
                print(type(err), max_retries)
                print(job.status())
                new_batch.append(batch[j])
        batch = new_batch
        max_retries -= 1
        if max_retries <= 8:
            flag = 1
            while (flag):
                try:
                    client = Client("http://demo.labinform.ru:30100/")
                    flag = 0
                except Exception as err:
                    print(type(err), " server")
                    flag = 1
```
